```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
 
y,sr=librosa.load('/home/linhui.dai/pytorch_spectrogram/data/ravdess/Actor_01/03-01-01-01-01-01-01.wav', duration=2.97)
print(y.shape)
print(sr)
 
plt.plot(y)
```

```python
chroma_stft = librosa.feature.chroma_stft(y=y, sr=sr,n_chroma=12, n_fft=4096)
# plt.figure(figsize=(15,15))
# plt.subplot(3,1,1)
librosa.display.specshow(chroma_stft, y_axis='chroma')
plt.title('chroma_stft')
plt.colorbar()
```

```python
chroma_cq = librosa.feature.chroma_cqt(y=y, sr=sr)


librosa.display.specshow(chroma_cq, y_axis='chroma', x_axis='time')
plt.title('chroma_cqt')
plt.colorbar()
```

```python
chroma_cens = librosa.feature.chroma_cens(y=y, sr=sr)
print(chroma_cens.shape)
librosa.display.specshow(chroma_cens, y_axis='chroma', x_axis='time')
plt.title('chroma_cens')
plt.colorbar()
```

```python
S = librosa.feature.melspectrogram(y, sr=sr, n_mels=128)
log_S = librosa.amplitude_to_db(S, ref=np.max)
# log_S = librosa.logamplitude(S, ref_power=np.max)


librosa.display.specshow(log_S, sr=sr, x_axis='time', y_axis='mel')

# Make the figure layout compact
plt.tight_layout()
print(log_S.shape)
```

```python
# Next, we'll extract the first 128 Mel-frequency cepstral coefficients (MFCCs) 
# to make the size consistent with the spectrogram
mfcc = librosa.feature.mfcc(S=log_S, n_mfcc=128)
# delta and second delta of mfcc can show more time domain information
delta_mfcc  = librosa.feature.delta(mfcc)
delta2_mfcc = librosa.feature.delta(mfcc, order=2)
```

```python
librosa.display.specshow(mfcc)
plt.ylabel('MFCC')
plt.colorbar()
# plt.show(mfcc)
print(mfcc.shape)
```

```python
librosa.display.specshow(delta_mfcc)
plt.ylabel('D-MFCC')
plt.colorbar()
```

```python
librosa.display.specshow(delta2_mfcc)
plt.ylabel('DD-MFCC')
plt.colorbar()
```
